```python
import pandas as pd
```

```python
data = pd.read_csv('./INPUT/master.csv')
data
```

```python
data.columns
```

```python
data = data.drop(['country-year', 'HDI for year', 'generation'], axis = 1)
data
```

```python
paisesAelegir = ['Denmark','Finland','Sweden','Norway', 'Iceland', 'Spain']
```

```python
df_listo = data[(data['country'].isin(paisesAelegir))]
df_listo
```

```python
df_listo['sex'].value_counts()
```

```python
df_sex = pd.get_dummies(df_listo['sex'])
df_sex
```

```python
df_listo = pd.concat([df_listo, df_sex], axis=1)
df_listo
```

```python
#BORRAR COLUMNA SEXO
df_listo = df_listo.drop(['sex'], axis=1)
df_listo
```

```python
# RENAME COLUMNS
df_listo = df_listo.rename(columns = {'country': 'Country', 'year': 'Year', 'age': 'Age', 'suicides_no': 'Total_suicides','female': 'Female_suicides', 'male': 'Male_suicides'})
df_listo
```

```python
# REORDENAR COLUMNAS
df_listo = df_listo[['Country', 'Year', 'Age', 'Male_suicides', 'Female_suicides', 'Total_suicides', 'population','suicides/100k pop', ' gdp_for_year ($) ', 'gdp_per_capita ($)']]
df_listo
```

```python
df_listo['Male_suicides'] = df_listo['Male_suicides']*df_listo['Total_suicides']
df_listo['Female_suicides'] = df_listo['Female_suicides']*df_listo['Total_suicides']
df_listo
```

```python
df_listo.dtypes
```

```python
df_listo[' gdp_for_year ($) '] = df_listo[' gdp_for_year ($) '].apply(lambda x: x.replace(",", ""))
df_listo
```

```python
df_listo[' gdp_for_year ($) '] = df_listo[' gdp_for_year ($) '].astype(int)
```

```python
df_listo.dtypes
```

```python
df_listo = round(df_listo.groupby(['Country','Year']).mean(), 2)
```
